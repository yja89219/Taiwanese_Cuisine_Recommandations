```python
'''
Data cleaning and using Jieba Library to get top 20 keywords
'''
import numpy as np
import pandas as pd

food_df = pd.read_csv('PTT_food01.csv', header = None)
food_df
```

```python
clean_string_list = []
for i in range(len(food_df.index)):
    #selecting the cell
    step1 = food_df.iloc[i][6]
    #Cutting off the head
    step2= step1.split('\n───────────────────────────────────────\n')[1]
    #Cutting off the tail
    step3 = step2.split('※ 發信站:')[0]
    step4 = step3.split('\n')
    step5 = ''.join(step4)
    clean_string = step5.rstrip('--')
    clean_string_list.append(clean_string)

clean_string_sum = ''.join(clean_string_list)
clean_string_sum
```

```python
import jieba.analyse
```

```python
#TextRank analysis
top20_topics = jieba.analyse.textrank(clean_string_sum, topK=20, withWeight=False, allowPOS=('ns', 'n', 'vn', 'v'))
print(top20_topics)
```

```python
#TF-IDF analysis
top20_topics_2 = jieba.analyse.extract_tags(clean_string_sum, topK=20, withWeight=False, allowPOS=('ns', 'n', 'vn', 'v'))
print(top20_topics_2)
```
